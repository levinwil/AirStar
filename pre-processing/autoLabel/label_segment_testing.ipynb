```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./test.png', 0)
```

```python
plt.figure()
plt.imshow(img)
plt.show()
```

```python
maxArea = 0
maxLabel = 0
for i in range(1, np.max(img)):
    area = len(zip(*np.where(img == i)))
    if area > maxArea:
        maxArea = area
        maxLabel = i
        
print maxLabel
```

```python
plt.figure()
plt.imshow(img == maxLabel)
plt.show()
```
